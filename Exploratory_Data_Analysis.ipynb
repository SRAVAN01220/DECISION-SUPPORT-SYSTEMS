```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load cleaned dataset
df = pd.read_csv('Cleaned_StudentsPerformance.csv')

# View first 5 rows
df.head()
```

```python
# Basic statistics
df.describe()
```

```python
# Histogram of scores
plt.figure(figsize=(12,4))
for i, column in enumerate(['math score', 'reading score', 'writing score']):
    plt.subplot(1,3,i+1)
    sns.histplot(df[column], kde=True, color='skyblue')
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# Bar plot for Pass/Fail
df['pass_fail_Pass'] = df['pass_fail_Pass'].astype('category')
sns.countplot(x='pass_fail_Pass', data=df, palette='Set2')
plt.title('Pass vs Fail Count')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplot (if 'gender_male' exists after encoding)
if 'gender_male' in df.columns:
    df['gender'] = df['gender_male'].apply(lambda x: 'Male' if x==1 else 'Female')
    plt.figure(figsize=(8,5))
    sns.boxplot(x='gender', y='average_score', data=df)
    plt.title('Average Score by Gender')
    plt.show()
```
